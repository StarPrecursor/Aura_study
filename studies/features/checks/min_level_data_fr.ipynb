## Funding rate minutes data

```python
symbol = "BTCUSDT"
features = ["timestamp", "funding_timestamp", "funding_rate", "index_price", "mark_price"]
date_start = datetime.datetime(2022, 1, 1)
date_end = datetime.datetime(2022, 9, 1)
df_fr = toy.toy_data.get_symbol_data_coarse(symbol, date_start, date_end, features, use_8h=False)
df_fr["time"] = pd.to_datetime(df_fr["timestamp"], unit="us")
df_fr = df_fr.sort_values(by="time")
df_fr.head()
```

```python
# resample to 1min period
df_fr = df_fr.resample("1min", on="time").mean()
df_fr = df_fr.reset_index()
df_fr.head()
```

```python
# Add rolling features
new_cols = {}
for lookback in [3, 5, 10, 20, 50]:  # unit: 8h
    lb_min = lookback * 8 * 60
    new_cols[f"funding_rate_rol_mean_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).mean()
    )
    new_cols[f"funding_rate_rol_std_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).std()
    )
    new_cols[f"funding_rate_rol_max_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).max()
    )
    new_cols[f"funding_rate_rol_min_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).min()
    )
    new_cols[f"funding_rate_rol_skew_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).skew()
    )
    new_cols[f"funding_rate_rol_kurt_{lookback}"] = (
        df_fr["funding_rate"].rolling(lb_min, min_periods=1).kurt()
    )
df_fr_append = pd.DataFrame(new_cols)
df_fr = pd.concat([df_fr, df_fr_append], axis=1)
```

```python
agg_dict = {
    "funding_rate": ["mean", "std", "max", "min", "skew", "last"],
}

df_fr["time_H"] = df_fr["time"].dt.floor("H")
for hl in [5, 30, 60, 240, 1440]:  # in minutes
    df_fr[f"funding_rate_ema_{hl}"] = df_fr["funding_rate"].ewm(span=hl).mean()
    agg_dict[f"funding_rate_ema_{hl}"] = "last"
df_fr_agg = df_fr.groupby("time_H").agg(agg_dict)
df_fr_agg = df_fr_agg.shift(1)
df_fr_agg.columns = ["__".join(x) for x in df_fr_agg.columns.ravel()]
df_fr_agg.reset_index(inplace=True)
df_fr_agg.head()
```

```python
# add future funding rate cumsum
new_cols = {}
for lookforward in [1, 3, 5, 10]:  # unit: 8h
    lf_h = lookforward * 8
    new_cols[f"funding_rate_future_{lookforward}"] = (
        df_fr_agg["funding_rate__last"].rolling(lf_h).sum().shift(-lf_h)
    ) / 8
df_fr_append = pd.DataFrame(new_cols)
df_fr_agg = pd.concat([df_fr_agg, df_fr_append], axis=1)
df_fr_agg.head()
```
